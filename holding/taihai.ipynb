```python
HIGH = data_day_qfq.high
LOW = data_day_qfq.low
CLOSE = data_day_qfq.close
rolling_atr = ta.ATR(HIGH, LOW, CLOSE, timeperiod=90).shift(1)
rolling_atr_divide_close = rolling_atr / CLOSE.shift(1)

ind_atr_pos = (open_pre_close_diff.abs() > rolling_atr_divide_close) & (open_pre_close_diff >= 0)
ind_atr_neg = (open_pre_close_diff.abs() > rolling_atr_divide_close) & (open_pre_close_diff < 0)

```

```python
data_day_qfq.data[ind_atr_neg]
```

```python
((data_day_qfq.data[ind_atr_neg].close - data_day_qfq.data[ind_atr_neg].open) / \
(data_day_qfq.data[ind_atr_neg].open)).sum()
```
