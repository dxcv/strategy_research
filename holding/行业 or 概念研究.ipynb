```python
def return_series(dataframe):
    CLOSE = dataframe.close
    PCT_CHANGE = CLOSE.pct_change()
    res = (1+PCT_CHANGE).cumprod()
    return pd.DataFrame({'CUM_RETURN':res})
```

```python
ind = data_hedian.add_func(return_series)
```
