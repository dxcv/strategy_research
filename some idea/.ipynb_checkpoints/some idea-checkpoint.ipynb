实况神圣年终2015年投资经验汇总【个股只是举例不构成买入建议】

一：构建交易系统
做股票，必须有自己的一套系统，
比如仓位配置，选股倾向，擅长领域，
风险控制模式等，严格按照自己设置的止盈止损方法操作，

并且用自己长处做股票，
比如你是某企业的员工，可能你对你企业文化和未来的发展模式和核心技术，东西的质量就了解;
你是做销售的可能你对你销售的东西的销量，市场占有率就了解;
有的人长期研究趋势，可能他对什么时候破位止损，盘感方面就更擅长;
有的人宏观研究的好，他可能对未来什么行业景气度高更了解;
有的人控制仓位厉害，可在行情好时捕捉热点连续大赚，行情差时他会控制仓位控制风险。

把自己擅长的领域相对重仓些是比较好的办法。
比如有的人擅长捕捉热点，有的人擅长分析基本面研究未来的发展情况预期情况，分析企业的内在价值，
有的人擅长技术面，研究趋势，研究资金的关注程度，
有的人擅长消息面，能够及时洞察到影响股票的信息并对信息做出敏锐的解读做出自己的及时判断，

同样每个人止损止盈的设计也不同，
有的人是趋势走坏了就果断止损，
有的人是高位滞胀了连续不再持续上升就选择其他强势股换仓，
有的人是仓位控制强项，弱势的时候会减仓控制仓位行情好的时候重仓出击，
有的人擅长基本面当业绩低于预期或者基本面发生改变就会止损，企业业绩有明显改善被及时洞察到提前潜伏。有的是深入研究企业文化，企业发展模式，大环境的影响，对企业未来进行分析，那么未来没有达到预期就会止损。

只说技术面，也有每个人擅长的，
比如有的人擅长买卖点的分析，
有的人擅长分析趋势，
有的人擅长关注成交量的变化，
有的人擅长观察上升通道，
有的人对圆弧底观察的较多，
有的人擅长强势股的捕捉，
有的人擅长波浪理论，
有的人擅长缠论，
有的人擅长杯柄，

总之必须结合自己擅长的领域总结一套适合自己的交易方式严格按照交易规则操作，不因大盘的短期波动而影响自己客观分析市场变化。

严格按照自己的系统操作，
不断发现自己系统中存在的漏洞不断完善，并且每天复盘，
选择一些未来看好的股票作为观察不断看自己选择的股票是否达到自己选择的目标，不断提高自己的盘感，长期下去会不断的提高进步。

具体到操作上，  
首先我们股市波动偏大，所以我的选择是大多数时间不随意出手轻易重仓，热点开始增加行情开始转好大盘成交量开始逐步增加涨停家数开始增多的时候，资金参与逐步增加。
其次就是大部分股票市盈率确实很低，也很多人认为做低市盈率的股票安全，
但长期看低市盈率的表现并不一定有成长股涨幅喜人，
这个和资金面有关，
和大家对未来发展方向政策导向的预期也有关，
所以做低市盈率的最好也做业绩持续性好的未来业绩增长预期强烈的，这样业绩继续预增高成长也还是会有一定的超额收益，如我2015年11月到12月做的002019亿帆鑫富，还有如美的集团这类。苏宁云商涨的最好的阶段就是从小企业慢慢变大而且不断高送转的那段而不是做大做强以后。昆明制药，网宿科技这些涨的最快的那段也是从小变大的过程。而真正已经做大做强的大企业很多每年涨幅都很难超过40%

`仓位管理`
我个人不喜欢集中1只，集中持股固然在牛市里也许可以收益最大化，但风险也很大，如果碰见重庆啤酒这样的或者2015年6月的停牌股，如果当时停牌就无法及时止损。突发事件在A股很多，且有的公司信息披露不透明，那么分散持股相对可以降低对单一品种的风险，

但分散持股不是随便买N多股票，也是要合理
分配长线短线权重，依然要在几千只股票里精选细选个股，

就长线选股而言，我喜欢选择有未来发展空间的，业绩持续性好的，题材多的且题材都是大题材为主的，每当有某个大题材符合的时候都会短期有波上涨行情的。

中线部分做题材预期没有兑现的。当然任何长线股如果短期涨幅过大我都会减仓1半留底仓，这样剩下的跌回成本的几率降低，也更容易拿得住股票。
中线的仓位可以以技术面结合题材预期考虑多一些，是否底部拐头向上同时又有短期的大的题材预期呼应，且题材预期还没有兑现，技术面每个人爱好不同能力不同，我中线题材预期的股喜欢题材兑现后技术面如果放量高开低走而前期已经有很高涨幅就选择减仓或清仓。

短线行情特别好的时候做强势涨停开板票小部分参与，
部分做短期爆发性题材的,这类持股因为本身就是短线操作，走弱即可卖出，题材不持续了也可以马上卖出。

一般原则上长线品种很少操作，短线品种在行情好的时候更多操作，但盘感不好的时候应该减少操作频率。
部分长线部分短线的好处是如果长线判断未来基本面出现了问题或者大盘整体中期弱势拖累个股长期的收益的时候，如果自己短线仓位有不错的阶段性收益，也可以对总资金有正面的影响。

分散持股只要大部分选择的不错比如3只长线票里3月内2只涨30%1只跌20%就是很不错收益，
短线持仓里如果10次操作7次正收益就是有一定的正面贡献。

需要说明的是，我个人在对大盘把握不好的时候会选择仓位控制，这样也是降低风险减少回撤的一个办法。

每天选几只觉得以后会涨的放自选，标记好技术面资金面或者基本面的选择理由，长线还是短线。
然后到时候看涨了没有，来检验自己进步没有，如果每次选的自选股后来都有不错表现就是进步了，如果表现的都很不理想就要分析自己的逻辑性和技术面哪些地方不足不断完善他。

这段时间(文章写于2015年12月28日)建议就不要仓位太重，因为失误可能是比较高的一段或者盘感是比较差的一段。每天选择一些典型案例放入观察盘并且观察一些重点消息每日进行解读后面看市场的表现来检验自己和市场的距离。

我个人不喜欢全部投资A股，可以部分资金低风险的如货币基金，部分资金参与申购新股，部分资金可以逆回购，部分资金可以投资其他市场如港股或者美股基金，在有能力的也可以布局点美股，部分资金可以投资一些资金看好的或者喜欢的实物。分散在不同的地方可以降低风险。当然根据不同时期不同的环境，不断调整投资各种领域的权重。

谈成长股我更喜欢盘子小一些的但发展空间比较大的，盘子过大的企业一是要上涨需要很多的资金推动，
2是如果已经市场占有率很大了再不断扩展的难度就加大了，比如1家面馆扩展到2家面馆容易，1个宝钢这么大的规模的企业变成2个宝钢那么大的企业的规模难度就大的多。

另外成长股还要看未来发展空间，举个例子，雾霾天气的时候空气净化器的需求大增，而最早的阶段空气净化器的厂商还不多，这段时间这类企业的股价就会上涨的比较多，但假如有一天雾霾天气好转，更多企业都开始做空气净化器了，供需关系发生了改变，这类股票我就会选择规避。

下面谈的长线中线主要进行举例说明，我个倾向于业绩持续性好的成长股和政策长期支持的大的题材股，选择题材股也最好选择题材多的股票，这样不管碰见哪个题材都可能涨一波。

比如2015年的张江高科，他又是迪斯尼，又是上海自贸区，还是创投题材，还是上海科创板题材，降息降准还利好房地产板块，还是上海国资改革题材。这类题材股碰见任何利好都可能有一波上涨可能，

说完长线举例中线，中线选择的一般就是短期预期，比如2015年博鳌论坛开幕前做海南股，广东自贸区挂牌前2015年的4月21号前做广东自贸区自贸区题材股。冬奥会申奥前2015年7月底做体育概念股等。当然牛市选择长线股还可以多选择低价小盘股，这类也是几轮牛市中最好的品种。【只局限在注册制前】
牛市还可以选B类基金上折预期强的增加短期收益，弱势则要布局A类基金下折预期强的对冲风险。

二：谈谈哪类消息容易见光死消息一般是突发消息比如局部冲突突然爆发或者哪里禽流感这类没有办法提前知道的消息 消息出现也可以持有，
2是超预期的消息，消息出现前没有大幅拉升的，消息兑现也可以继续持有。如果是如2015年的降息预期，降准预期，周末前提前市场反映了相关板块的，或者某个提前预期很久了比如广东自贸区或者上海自贸区，挂牌前已经轮番拉升了的，那么消息兑现就要卖出相关题材股。【参考2013年9月的外高桥和2015年4月的珠海港等】

三：如何选择预期个股更多时候还是要做预期的题材，就是还没有发生的但很快会发生的。如广东自贸区，挂牌前埋伏，挂牌了就出。最好做多题材的预期，如果一个企业很多题材都还是朦胧期。比如他又是地产可能周末大家预期降息降准。又是国企改革预期，又是合并预期，那么多预期共同存在的肯定比单一题材更吸引不同资金关注。
`题材也要给权重值,比如每只股票按照题材打分`
做预期也最好做大题材，就是高层政策支持力度大的预期，这类题材市场会更认可。举个例子：我过去周4做国企改革一般喜欢买中粮地产。好像2015年4月那次国企改革板块当天涨的不错我做了中粮地产，正好那次距离上次降准也有一段时间了。降准预期也很大了，周末降准后周一地产股都高开，我顺势出货了，那天好像其他国企改革概念股中粮屯河，中成股份那些都低开低走吧

四：如何增加盘感。我关注涨停分析很久，自己也长期汇总新闻题材，长期复牌和分析一些典型技术特点和特殊技术特点事后验证，新闻汇总的题材事后验证，涨停分析的龙头热点长期对比都可以增加盘感。
长期题材汇总还可以增加对个股题材的记忆，长期每次有利好预期基本脑子里都会知道哪些属于哪个题材。

五：如何做到买到龙头。一般题材出来的时候观察以前这类题材谁是龙头，当然如果短期前期的那个龙头技术面结合看不太好，可以把技术面的前期龙2和前期龙1一起买还可以关注一些本次板块里最先启动的个股，这样一定可以整体提高收益稳定性。

六，如何提前挖掘公司题材：投资者互动平台可以了解很多企业的题材。
七：复牌什么情况涨什么情况跌大盘在高位的时候，比如5000点停牌的2800点复牌，哪怕重组的内容不错，也有补跌的可能性，如电广传媒2015年6月停牌，复牌后就补跌。大盘在低位，比如2800点停牌的票，大盘反弹到3700点复牌的，很多重组失败的也会补涨，比如好多3400点7月初停牌的，7月底3800复牌的都补涨了几个涨停。其他重组股票大盘波动不大的时候则更关注重组的内容，如果重组后增加了很大的利润，则大涨的概率更大。

`比如过去N天，当题材切换时(怎么界定题材切换), 资金流入多少来打分`
八，自选股观察盘：创业板关注乐视网，同花顺和东方财富。中小板关注苏宁云商。主板关注中信证券，中国平安同时关注工商银行和中国石油，来分析板块热点的切换。

同时买小盘股题材股的关注IC走势买大盘股的关注IH走势。关注涨停家数。这类指标只作为参考盘面用，我不会买入。【中国石油突然急拉过大对小盘股构成的短期压力不容小视】
`这里的IC和IH是股指期货吗`

九：选龙头的几个方法过去涨停次数多的，资金一般熟悉 喜欢关注这次最先奔涨停的，买入后封住涨停不开板的抛压小涨停后封单很大不撤单的，证明封单的买入意愿浓厚建议不要买涨幅最小的题材股，这类一般是没有资金关注，没有资金关注的最多和大盘同步

十：仓位分配的安全性最好不要把亏损股的借壳预期股重仓1只，风险永远放第一。而业绩好的成长股题材又多的可以稍微重仓点，因为这类股票就算出现短期震荡迟早也会受到资金关注。同样不要集中持有短线强势股防止一旦大盘
短期出现大幅波动这类股的波动也会很大，技术水平差的可能就会出现追涨杀跌。分散持股绝对不是买指数，而是要买到都非常看好的品种里，这样大部分持股收益很高就可以长期赚钱了。


其他交易心得：1：利好兑现前大涨的利好兑现开盘高开低走放量的要出，利好兑现前没有异动的，利好兑现开盘高走的就可以留
2：次新股里的未分配利润高的或者公积金高的比大盘股高送转概率更大
3：降息预期一定提前1天周四买，因为周5大家都做降息预期拉高就可以借机减仓点了。
